```python
from sklearn.model_selection import train_test_split
```

```python
X = data.iloc[:,0:78]
```

```python
X.columns
```

```python
Y = data.loc[:,['Label']]
Y.head()
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(X,Y,test_size=0.2, random_state=100, shuffle=True)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_tree = RandomForestClassifier(max_depth = 10, min_samples_split=40000, min_samples_leaf = 4000,random_state=100)
```

```python
rf_tree.fit(xtrain,ytrain)
```

```python
pred = rf_tree.predict(xtest)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(ytest,pred))
```

```python
rf_tree.feature_importances_
```

```python
features_tuple = list(zip(X.columns, rf_tree.feature_importances_))
```

```python
feature_imp = pd.DataFrame(features_tuple, columns=[
                           "Feature Names", "Importance"])
```

```python
feature_imp = feature_imp.sort_values("Importance", ascending=False)
```

```python
f = plt.figure()
f.set_figwidth(30)
f.set_figheight(20)
sns.barplot(x="Feature Names", y="Importance", data=feature_imp, color='b')
plt.xlabel("NIDS Features")
plt.ylabel("Importance")
plt.xticks(rotation=90)
plt.title("Decision Classifier - Features Importance")
```
